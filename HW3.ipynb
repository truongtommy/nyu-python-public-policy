```
#Read in cleaned 311 data 
df = pd.read_csv('/content/drive/My Drive/Data for Python/cleaned_311_data_hw3.csv', header='infer')
df.head(20)
```

```
#Create function that locates the word "graffiti" in the complaint type and descriptor columns
def graffiti_locator(row):
  if 'Graffiti' in row.complaint_type and pd.notnull(row.descriptor):
    return True
  else: 
    return False 
```

```
#Check if type is function
type(graffiti_locator)
```

```
#Apply function above to the df 
df['graffiti_flag']= df.apply(graffiti_locator,axis=1)
df
```

```
#Check if the df has both TRUE and FALSE values 
check_1 = df.groupby('graffiti_flag').size()
check_1
```

```
#Create another df that only contains records where graffiti_type is TRUE 
df_graffiti = df[df.graffiti_flag]
df_graffiti
```

```
#Group by count of requests per community board 
df_graffiti.groupby('community_board').size().nlargest(15).reset_index(name='count')
```
