```
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np
```

```
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
drive.mount('/content/drive')
```

```
df = pd.read_csv('/content/drive/My Drive/Data for Python/cleaned_311_data_revised.csv', header='infer')
df.head()
```

```
top_10 = df.groupby('complaint_type').size().nlargest(10).reset_index(name='count_of_311_requests').sort_values('count_of_311_requests', ascending=False)
top_10
```

```
filtered_311 = df[df.complaint_type.isin(top_10.complaint_type)]
filtered_311.head(10)
```

```
filtered_311.created_date = pd.to_datetime(filtered_311.created_date, format='%m/%d/%Y %I:%M:%S %p')
```

```
filtered_311.head()
```

```
filtered_311['year'] = filtered_311.created_date.dt.year
filtered_311['date'] = filtered_311.created_date.dt.date
filtered_311['month'] = filtered_311.created_date.dt.month
filtered_311['weekday_name'] = filtered_311.created_date.dt.weekday_name
filtered_311['weekday'] = filtered_311.created_date.dt.weekday
filtered_311['hour'] = filtered_311.created_date.dt.hour

filtered_311.head()
```

```
per_hour = filtered_311[['hour','date','complaint_type']].groupby(['hour','date','complaint_type']).size().reset_index(name='count_requests')
 
per_hour.head()
```

```
per_hour_median = per_hour.groupby(['hour','complaint_type'])['count_requests'].median().reset_index(name='median_requests')

per_hour_median.head()
```

```
plt.figure(figsize=(20, 6))
for type in top_10.complaint_type:
  filtered = per_hour_median[per_hour_median.complaint_type == type]
  plt.plot(filtered.hour, filtered.median_requests,label = type)
  plt.legend()
plt.grid()
plt.title('Count of Complaints per Hour')
plt.xlabel('Hour')
plt.ylabel('Count of 311 Requests')
```

```
per_weekday = filtered_311[['weekday','date','complaint_type']].groupby(['weekday','date','complaint_type']).size().reset_index(name='count_requests')
 
per_weekday.head()
```

```
per_weekday_median = per_weekday.groupby(['weekday','complaint_type'])['count_requests'].median().reset_index(name='median_requests')

per_weekday_median.head()
```

```
plt.figure(figsize=(20, 6))
for type in top_10.complaint_type:
  filtered = per_weekday_median[per_weekday_median.complaint_type == type]
  plt.plot(filtered.weekday, filtered.median_requests, label = type)
  plt.legend(loc=2)
plt.grid()
plt.title('Count of Complaints per Weekday')
plt.xlabel('Weekday')
plt.ylabel('Count of 311 Requests')
```

```
per_month = filtered_311[['month','year','complaint_type']].groupby(['month','year','complaint_type']).size().reset_index(name='count_requests')
 
per_month.head()
```

```
per_month_median = per_month.groupby(['month','complaint_type'])['count_requests'].median().reset_index(name='median_requests')

per_month_median.head()
```

```
plt.figure(figsize=(20, 6))
for type in top_10.complaint_type:
  filtered = per_month_median[per_month_median.complaint_type == type]
  plt.plot(filtered.month, filtered.median_requests, label = type)
  plt.legend(loc=2)
plt.grid()
plt.title('Count of Complaints per Month')
plt.xlabel('Month')
plt.ylabel('Count of 311 Requests')
```

Most types of noise complaints seem to occur at a higher volume either late at night or early in the morning during the weekend. Heat and hot water complaints happen in the colder months more (Jan, Feb, Dec). Bulky item collection and illegal parking complaints seem to occur more during the middle of the day.
